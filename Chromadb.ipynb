```python
import chromadb
client = chromadb.Client()
collection = client.create_collection(name="new_collection")
```

```python
collection.add(
    documents=[
        "This is about Indian dishes",
        "This is about Italian dishes"
    ],
    
    ids = ['id1','id2']
)
```

```python
results = collection.query(
    query_texts=['Query is about pasta dishes'],
    n_results=1
)
results
```
